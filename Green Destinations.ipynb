#### `Some of the Features are:`

`Age:` The age of the employee.<br>
`Attrition:` Indicates whether the employee has left the company (Yes) or is still employed (No).<br>
`BusinessTravel:` Frequency of the employee's business travel, typically categorized as 'Non-Travel', 'Travel_Rarely', or 'Travel_Frequently'.<br>

`PercentSalaryHike:` The percentage increase in the employee's salary compared to the previous year.<br>
`MonthlyIncome:` The monthly salary of the employee.<br>
`YearsAtCompany:` The number of years the employee has worked at the current company.

## `Calculate the Attrition Percentage`

```python
# Calculate the total number of employees
total_employees = len(gd_df)

# Calculate the number of employees who left
attrition_count = gd_df[gd_df['Attrition'] == 'Yes'].shape[0]

# Calculate the percentage of attrition
attrition_percentage = (attrition_count / total_employees) * 100

# Display the result
print(f"Attrition Percentage: {attrition_percentage:.2f}%")
```

## `Visualize the Attrition Percentage`

```python
# Data for pie chart
attrition_labels = ['Stayed', 'Left']
attrition_values = [total_employees - attrition_count, attrition_count]

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(attrition_values, labels=attrition_labels, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], startangle=90)
plt.title('Employee Attrition Percentage')
plt.show()
```

```python
# Select relevant features
relevant_features = ['Age', 'YearsAtCompany', 'MonthlyIncome', 'Attrition']

# Calculate the correlation matrix
correlation_matrix = df[relevant_features].corr()

# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
